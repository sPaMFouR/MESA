# Introduction

- Modules for Experimentation in Stellar Atrophysics (MESA)
- 1D Stellar Evolution Code
- Runs on less-powered laptops to parallelised multi-core Clusters
- Open-Source, wide applications and friendly to user modifications
 - Various Environments
 - Exploration of stars from any region on HR diagram
 - Exotic Situations



# Installation

- MESA requires: <br>
    Mac or Linux operating system (Windows users may want to try MESA-Docker)<br>
    64-bit processor <br>
    8 GB RAM <br>
    20 GB free disk space <br>

- Download MESA Software Development Kit [[SDK]](http://www.astro.wisc.edu/~townsend/static.php?ref=mesasdk) (contains all the necessary compilers, libraries, etc.)

- Follow installation Steps [[Link]](https://docs.mesastar.org/en/latest/installation.html) (Super user permissions not needed)

# Controls

- Modules are in fortran (EOS, GYRE, KAP, etc.)
- __inlist__ files controls the behaviour, direct to other inlists. Actual science case is bbuild through inlists
- Five sections are there in __inlist__
 - star_job (Options for star evoluton) (nuclear reactions, metallicity, rot., etc.)
 - eos ( provides options for eos module)
 - kap (opacities) 
 - controls (winds, boundaries, stopping conditions, etc.)
 - pgstar (different plots, customised for runtime preview)

- Definition is of type
> name = value _! comment_
- Defaults are mentioned in mesa/__module__/defaults/__module__.defaults (use only those you wish to change) 
- Check what options can be used under various sections (some are on/off switches and some have values)


# Using Test Suites

- MESA provides number of example inlist which already have set of controls defined which we can change as per our problem
- Best to start from there (start from simple example case)

## We start with Core-Collapse SNe Test Suite

- First evolve star from pre-ZAMS to development of Iron Core using _make_pre_ccsn_typeIIp_ (Necessary changes in list?)
> ./clean ./mk ./rn ./re or ./re _photo_
- Can check history and detailed profiles through LOGS (Tables)
- Use output model from 1 to another test suite where the fe core is removed at some infall
- Energy(__inlist_edep__) and Ni(__xctrl(12)__ in __inlist_shock_part5__) are given (_ccsn_IIp_)
- __inlist_stella__ to get output as required by __STELLA__


# STELLA
- Functionality to evolve just before shockout to later phases
- Use output from _ccsn_IIp_ in stella/modmake
- Make changes in stella/strad/run/mesa.dat (Tcur)
- Check output from stella/res folder (vel., lum., magnitudes, etc.)
